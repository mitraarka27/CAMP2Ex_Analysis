```python
#!/usr/bin/python

#import necessary modules
from pyhdf import SD
import numpy as np
import math
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt
import MisrToolkit as Mtk
import sys
import glob
import h5py
import time
import calendar
import scipy
from scipy import spatial

#Multiple nearest-neighbour search routine
def do_kdtree_multi(A, pt):
    mytree = scipy.spatial.cKDTree(A,balanced_tree=False,compact_nodes=False)
    dist, indexes = mytree.query(pt,distance_upper_bound=np.inf, k=10)
    return dist, indexes

#Calculate physical distance between two geographical coordinates
def haversine(coord1, coord2):
    R = 6372800  # Earth radius in meters
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    phi1, phi2 = math.radians(lat1), math.radians(lat2) 
    dphi       = math.radians(lat2 - lat1)
    dlambda    = math.radians(lon2 - lon1)
    a = math.sin(dphi/2)**2 + \
        math.cos(phi1)*math.cos(phi2)*math.sin(dlambda/2)**2
    return 2*R*math.atan2(math.sqrt(a), math.sqrt(1 - a))

A = np.loadtxt('RF07_MISR/RF07_TerraOverpass_RSP_latlong')
rsp_lat=A[:,0]
rsp_lon=A[:,1]

file = 'RF07_MISR/MISR_AM1_TC_CLOUD_P116_O104919_F01_0001.hdf'
m = Mtk.MtkFile(file)
path_number = m.attr_get('Path_number')
r = Mtk.MtkRegion(path_number, 75,77)
d1 = m.grid('Stereo_WithoutWindCorrection_1.1_km').field('CloudTopHeight_WithoutWindCorrection').read(r)
d1 = d1.data()*0.001
d1 = d1.ravel()
d2 = m.grid('Stereo_1.1_km').field('CloudTopHeight').read(r)
d2 = d2.data()*0.001
d2 = d2.ravel()
map_info = r.snap_to_grid(path_number, 1110)
latlon = np.asarray(map_info.create_latlon())
lat = latlon[0,:,:]
lon = latlon[1,:,:]
misr_points = ([lat.ravel(), lon.ravel()])
misr_lat = []
misr_lon = []
misr_withoutwind_ht = []
misr_withwind_ht = []
flattened_point = []
for i in range(len(rsp_lat)):
    with_wind=0.;without_wind=0.;c_with=0;c_without=0;
    rsp_coord = [rsp_lat[i],rsp_lon[i]]
    dist, CM_collocation = do_kdtree_multi(np.transpose(misr_points),rsp_coord)
    flattened_point = np.append(flattened_point, CM_collocation[0])
    for n in range(0, len(CM_collocation)):
        nn_MISR_lat = misr_points[0][CM_collocation[n]]
        nn_MISR_long = misr_points[1][CM_collocation[n]]
        nn_MISR = [nn_MISR_lat,nn_MISR_long]
        if (haversine(nn_MISR,rsp_coord)/1000.<=5.):
            d1_col = d1[CM_collocation[n]]
            d2_col = d2[CM_collocation[n]]
#             print(i,n,d1_col,d2_col)
            if (d1_col>0.):
                without_wind = without_wind+d1_col
                c_without = c_without+1
            if (d2_col>0.):
                with_wind = with_wind+d1_col
                c_with = c_with+1  
    if (c_without>1):
        without_wind = without_wind/c_without
    else:
        without_wind = -9.999
    if (c_with>1):
        with_wind = with_wind/c_with
    else:
        with_wind = -9.999
    misr_lat = np.append(misr_lat, misr_points[0][CM_collocation[0]])
    misr_lon = np.append(misr_lon, misr_points[1][CM_collocation[0]])
    misr_withoutwind_ht = np.append(misr_withoutwind_ht, without_wind)
    misr_withwind_ht = np.append(misr_withwind_ht, with_wind)
    print(i,with_wind,without_wind)
```
